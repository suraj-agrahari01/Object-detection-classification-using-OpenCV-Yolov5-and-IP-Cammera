# live web cam detction using yolov5

```python
#yolo_web_cam_video

import cv2
import torch
import numpy as np

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # You can specify the model size (e.g., 'yolov5s', 'yolov5m', 'yolov5l', 'yolov5x')

# Set the model to evaluation mode
model.eval()

# Initialize the video capture
cap = cv2.VideoCapture(0)  # Use the appropriate video source (e.g., 0 for default camera)

while True:
    ret, frame = cap.read()  # Read a frame from the camera

    # Perform inference
    results = model(frame)

    # Get detected objects and their details
    detections = results.pred[0]
    for det in detections:
        label = model.names[int(det[-1])]
        confidence = det[4].item()
        bbox = det[:4].tolist()

        # Draw bounding boxes and labels
        if confidence > 0.5:  # You can adjust the confidence threshold as needed
            x1, y1, x2, y2 = map(int, bbox)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f'{label}: {confidence:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('YOLOv5 Object Detection', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
